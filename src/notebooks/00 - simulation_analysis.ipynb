```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
FOLDER_PATH = "../../results/"
with open(FOLDER_PATH + "single_simulation_log.pkl", "rb") as f:
    log = pickle.load(f)
```

```python
wind_gt  = np.asarray(log["wind_gt"])    # shape (T, 2)
wind_ekf = np.asarray(log["wind_ekf"])   # shape (T, 2)

T = wind_gt.shape[0]
dt = 0.01
time = np.arange(T) * dt

labels = ["x", "y"]

fig, axs = plt.subplots(2, 1, figsize=(10, 5), sharex=True)

for i in range(2):
    axs[i].plot(time, wind_gt[:, i], label="Wind GT")
    axs[i].plot(time, wind_ekf[:, i], "--", label="Wind EKF")
    axs[i].set_ylabel(f"w_{labels[i]} [m/s]")
    axs[i].grid(True)

axs[-1].set_xlabel("Time [s]")
axs[0].legend()
fig.suptitle("Wind estimation (horizontal): Ground Truth vs EKF")

plt.tight_layout()
plt.show()
```
